```python
#%%file c:\\users\\sachin\\desktop\\result.py
import os
def result(path,sep='\t'):
    fp = open(path)
    data = fp.read()
    fp.close()
    rows = data.split('\n')
    new_data = [] 
    for line in rows : 
        d = line.split(sep)
        d[2] = int(d[2])
        d[3] = int(d[3])
        d[4] = int(d[4])
        d.append((d[2]+d[3]+d[4])/3)
        d[2] = str(d[2])
        d[3] = str(d[3])
        d[4] = str(d[4])
        d[5] = str(round(d[5],2))
        d = '\t'.join(d)
        new_data.append(d)
    #print(new_data)
    final_data = '\n'.join(new_data)
    #print(repr(final_data))
    dir_name = os.path.dirname(path)
    fname = path.split('\\')[-1]
    fname = fname.split('.')
    fname = fname[0]+'_result.'+fname[1]
    fname = os.path.join(dir_name,fname)
    fp = open(fname,'w')
    fp.write(final_data)
    fp.close()
    print("Your Result is sucessfully written")
```

```python
#%%file c:\\users\\sachin\\desktop\\result.py
import os
def result(path,sep='\t'):
    fp = open(path)
    data = fp.read()
    fp.close()
    rows = data.split('\n')
    new_data = [] 
    for line in rows : 
        d = line.split(sep)
        d.append(str(round(sum(list(map(eval,d[2:])))/(len(d)-2),2)))
        d = '\t'.join(d)
        new_data.append(d)
    final_data = '\n'.join(new_data)
    dir_name = os.path.dirname(path)
    fname = path.split('\\')[-1]
    fname = fname.split('.')
    fname = fname[0]+'_result.'+fname[1]
    fname = os.path.join(dir_name,fname)
    fp = open(fname,'w')
    fp.write(final_data)
    fp.close()
    print("Your Result is sucessfully written")
```
